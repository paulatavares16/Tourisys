```python
import json
listChanged = []
with open('./data/brasil_reviews.json', 'r') as reader:
    with open('./data/map-reviews-cat.json', 'r') as mapReader:
        with open('./data/brasil_reviews_new_cat.json', 'w') as resultReader:
            mapCat = json.load(mapReader)
            for obj in reader:
                if(obj.strip() != ""):
                    item = eval(obj)
                    categories = list(item['categories'] or [])
                    catChanged = set(map(lambda cat: mapCat.get(cat), categories))
                    catChangedNotNull = filter(lambda cat: cat is not None, catChanged)
                    item['categories'] = catChangedNotNull
                    listChanged.append(item)
            resultReader.write(json.dumps(listChanged))
```
